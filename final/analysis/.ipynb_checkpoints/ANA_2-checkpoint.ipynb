#### TAKING USER INPUT FOR CITY 

```python
#User can enter any city name available in Zomato
# for eg.mumbai
input_var = input("Enter CITY: ")
print (input_var) 
```

#### STORING ALL THE JSON FILES DETAILS IN A LIST FOR THE CITY ENTERED BY USER

```python
temp =[] # STORING ALL THE JSON FILES DETAILS IN A LIST FOR THE CITY ENTERED BY USER

r= 0 # To keep count of the total number restaurants in that user mentioned city

cur_path = "/Users/sonalichaudhari/Desktop/final/data/"

for subdir, dirs, files in os.walk(cur_path):
        for file in files:
            if file.endswith('.json'): # Retrieving only the JSON files
                reqpath = subdir+'/'+file # Path of evry JSON file is stored in reqpath
                with open(reqpath) as json_file: # Opening the JSON file
                    d = json.load(json_file) # Loading the JSON file data in d
                    city = d['restaurant']['location']['city'].lower()  # Lowering the case of city name
                    
                    if city == input_var.lower(): # Checking if city name matches
                        temp.append(d)       # Appending the json to list temp
                        r=r+1
```

#### STORING THE TEMP LIST DATA IN A CSV

```python
with open('/Users/sonalichaudhari/Desktop/final/analysis/ana_2/ANA_2_PART1.csv', 'w') as csvfile:    
    fieldnames = ['RES_ID', 'Name','CUISINES','USER RATING','VOTES',\
                 'RATING TEXT','LATITUDE','LONGITUDE','AVG COST for TWO','CURRENCY']#Naming the title Row of the CSV
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for j in range(len(temp)-1):
        list_item =[]
        list_item = temp[j]['restaurant']['cuisines'].split(',') # Splitting the cuisines
        for x in range(len(list_item)-1):
            r = temp[j]['restaurant']['R']['res_id']
            n = temp[j]['restaurant']['name']
            cu= list_item[x].strip()
            u = temp[j]['restaurant']['user_rating']['aggregate_rating']
            v = temp[j]['restaurant']['user_rating']['votes']
            rt= temp[j]['restaurant']['user_rating']['rating_text']
            la= temp[j]['restaurant']['location']['latitude']
            lo= temp[j]['restaurant']['location']['longitude']
            avg=temp[j]['restaurant']['average_cost_for_two']
            cur=temp[j]['restaurant']['currency']
            writer.writerow({'RES_ID':r, 'Name':n,'CUISINES':cu,'USER RATING':u,'VOTES':v,\
                 'RATING TEXT':rt,'LATITUDE':la,'LONGITUDE':lo,'AVG COST for TWO':avg,'CURRENCY':cur}) 
```

#### CONVERTING THE CSV INTO DATAFRAME

```python
t = pd.read_csv('/Users/sonalichaudhari/Desktop/final/analysis/ana_2/ANA_2_PART1.csv').reset_index(0,drop=True)
t.head(3)
```

```python
#Sorting the dataframe t with user ratings and groupin by cuisines
```

```python
f = lambda x: x.sort('USER RATING', ascending=False)
p = t.groupby('CUISINES').apply(f).reset_index(0,drop=True)
s = p.groupby('CUISINES').head(1).reset_index(0,drop=True)

# For all cuisines availale in the user given city we got highest rated restaurant details.
s
```

```python
# Creating CSV of final output
```

```python
with open('ANA_2_PARTt2.csv', 'w') as csvfile:
    s.to_csv(csvfile, header=True)
```
